## Which events had students attended prior to coming to CCM? Show a bar chart of this field, sort from highest to lowest. Don't include an events which were not attended.

```python
events = ['a1_open_house', 'a2_instant_decision', 'a3_on_campus_info_session', 'a4_titans_tuesday',
        'a5_women_who_dare', 'a6_college_fair', 'a7_hs_sharetime_info',
        'a8_challenger_program', 'a9_cybersecurity_info_program', "a10_hs_info_session",
        'a11_hs_campus_visit', 'a12_solo_campus_visit', 'a13_worforce_development',
        'a14_corporate_training', 'a15_teen_arts_festival', 'a16_ccm_summer_camp']

events_count = newMajors[events].apply(lambda x: (x == 'Yes').sum())
events_count = events_count[events_count > 0]
events_count = events_count.sort_values(ascending=False)
events_count.plot(kind = 'bar', figsize = (10,6))

plt.title('Events Attended by Students Prior to Coming to CCM')
plt.xlabel('Event')
plt.ylabel('Number of Students')
plt.xticks(rotation=45, ha='right')
plt.show()
```

## Do those who changed majors from a different CCM program show more interest in taking computing classes?

```python
changed_majors = newMajors[newMajors['changed_majors'] != 'First Choice']
non_changed_majors = newMajors[newMajors['changed_majors'] == 'First Choice']

changed_majors_clean = changed_majors.dropna(subset=['interest'])
non_changed_majors_clean = non_changed_majors.dropna(subset=['interest'])

changedMajor_avgInterest = changed_majors_clean['interest'].mean()
nonChangedMajor_avgInterest = non_changed_majors_clean['interest'].mean()

print(f"Percentage of students who changed majors and are interested in computing classes: {changedMajor_avgInterest:.2f}%")
print(f"Percentage of students who did not change majors and are interested in computing classes: {nonChangedMajor_avgInterest:.2f}%")

interestComparison = pd.DataFrame({'Group': ['Changed Majors', 'Non-Changed Majors'],
                                  'Average Interest In Computing': [changedMajor_avgInterest, nonChangedMajor_avgInterest]})

interestComparison.plot(kind='bar', x='Group', y='Average Interest In Computing', legend=False, color=['skyblue', 'salmon'], figsize=(8, 6))
plt.title('Average Interest in Computing Classes: Changed Majors vs Non-Changed Majors')
plt.ylabel('Average Interest (1.0 to 5.0)')
plt.show()
```

## What is the ratio between students enrolled in a degree program versus certificate program?

```python
newMajors.loc[:, 'is_certificate'] = newMajors['student\'s_degree_program'].str.contains('certificate', case=False, na=False)
certificate_counts = newMajors['is_certificate'].value_counts()
certificate_counts.plot(kind='bar', color=['skyblue', 'salmon'], figsize=(8, 6))

plt.title('Number of Students with vs. without Certificate Programs')
plt.xlabel('Contains "Certificate"')
plt.ylabel('Number of Students')
plt.xticks(ticks=[0, 1], labels=['Without Certificate', 'With Certificate'], rotation=0)
plt.show()
```

## What is the top motivator for students seeking a computing degree/certificate at CCM?

```python
motivators = ['m1_job', "m2_bachelor's", 'm3_hs', 'm4_career_advancement', 'm5_career_change', 
             'm6_development', 'm7_job_displacement', 'm8_relocation', 'm9_keep_in_touch', 
            'm10_certifications', 'm11_finance', 'm12_enrichment']

motivator_counts = (newMajors[motivators] == 'Yes').sum()
motivator_counts = motivator_counts.sort_values(ascending=False)

plt.figure(figsize=(10, 6))
motivator_counts.plot(kind='bar', color='skyblue')
plt.title('Top Motivators for Students Seeking a Computing Degree/Certificate at CCM')
plt.xlabel('Motivators')
plt.ylabel('Number of "Yes" Responses')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

top_motivator = motivator_counts.idxmax()
print(f"The top motivator is: {top_motivator} with {motivator_counts.max()} 'Yes' responses.")
```
